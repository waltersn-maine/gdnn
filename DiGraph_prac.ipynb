```python
J = nx.DiGraph()
```

```python
J.add_node(1, elev=3)
J.add_node(2, elev=1)
J.add_node(3, elev=1)
J.add_node(4, elev=4)
J.add_node(5, elev=2)
J.add_node(6, elev=4)
```

```python
for i in range(1,7):
    for j in range(1,7):
        if i != j:
            if J.nodes[i]["elev"] > J.nodes[j]["elev"]:
                if J.has_edge(i,j):
                    J[i][j]["weight"]+=1
                else:
                    J.add_edge(i,j, weight=1)
            else:
                if J.has_edge(j,i):
                    J[j][i]["weight"] +=1
                else:
                    J.add_edge(j,i, weight=1)
```

```python
elabels = nx.get_edge_attributes(J, "weight")
```

```python
plt.figure(figsize=(3,3))
# define layout for nodes
pos = nx.spring_layout(J)
nx.draw(J, pos, with_labels=True, node_size=300, arrows=True)
nx.draw_networkx_edge_labels(J,pos,edge_labels=elabels)
plt.title("DiGraph practice")
plt.show()
```
